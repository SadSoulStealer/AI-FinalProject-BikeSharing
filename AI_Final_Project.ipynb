```python
#Importing the required libraries
from pycaret.utils import enable_colab
import numpy as np
import pandas as pd

#Include this statement in order to use PyCaret's functionalities
enable_colab()

#Importing dataset and gettng rid of the unneeded columns
df = pd.read_csv('hour.csv')
df.drop(['dteday','instant','casual','registered'], axis=1, inplace=True)

df.head()
```

Now we are going to split our Dataset to contain on set of data we will train on and one set of unseen data we will later use for predictions

```python
#Splitting data into two sets
df_data = df.sample(frac=0.9, random_state=786)
df_unseen = df.drop(df_data.index)

df_data.reset_index(drop=True, inplace=True)
df_unseen.reset_index(drop=True, inplace=True)

df.head()
```

Next, will be setting up the environment to take the data and preprocess it accordingly in order to be able to use this.

There are 3 important features to note after performing:

**Missing Values:** Displays True or False letting you know if you need to perform additional cleaning

**Categorical Variables:** Tells you how many you have

**Numeric Variables:** Tells you how many you have

```python
#Setting Up The Environment
from pycaret.regression import *
exp_AI = setup(data = df_data, target = 'cnt', session_id=123) 
```

Now that we have properly setup our data, we can run a comparison where PyCaret runs and and compares all the metrics of every model and lists them in a table for us

```python
#Comparing All Models
best = compare_models(exclude = ['ransac'])
```

Now we are going to create the top 3 models and compare them. PyCaret has an easy function for this named create_model we can use

```python
#Create Light Gradient Boosting Machine Model
lightgbm = create_model('lightgbm')
```

```python
#Create Extra Trees Classifier
et = create_model('et')
```

```python
#Create Random Forest Classifier
rf = create_model('rf')
```

Light Gradient Boosting Machine is the best model based on R2 and RMSE so we are going to perform the rest of the operations on it from here on.

Our next function is going to be tuning. Here we have tuned the lgbm model we have. There are different ways to tune each model.

```python
#Tune Light Gradient Boosting Machine
import numpy as np
lgbm_params = {'num_leaves': np.arange(10,200,10),
                        'max_depth': [int(x) for x in np.linspace(10, 110, num = 11)],
                        'learning_rate': np.arange(0.1,1,0.1)
                        }
tuned_lightgbm = tune_model(lightgbm, custom_grid = lgbm_params)
print(tuned_lightgbm)
```

Now that we  have tuned out model, we can produce the plots we need to look at. There are separate functions to produce all the plots separately, but we can run evaluate_model in order to create a interface to see them all

```python
#Evaluation Plots
evaluate_model(tuned_lightgbm)
```

Here we are just predicting the model and testing it

```python
#Predicting Model
predict_model(tuned_lightgbm);
```

Once you feel confident in the model you have created, you can run finalize_model in order to create the final version you plan to use for deployment

```python
#Finalizing Model
final_lightgbm = finalize_model(tuned_lightgbm)
print(final_lightgbm)
predict_model(final_lightgbm);
```

Here we run the final model we have created on our unseen data from earlier in order to test it. Also, you can use the check_metric to see how your R2 has improved from the first iteration

```python
#Predicting on the unseen data
from pycaret.utils import check_metric
unseen_predictions = predict_model(final_lightgbm, data=df_unseen)
unseen_predictions.head()

check_metric(unseen_predictions.cnt, unseen_predictions.Label, 'R2')
```
